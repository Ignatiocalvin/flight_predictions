```python
import pandas as pd

# read unprocessed data
df_unprocessed = pd.read_csv('data/5guys_flight_data.csv')
```

## Unprocessed

```python
# histogram of every column
df_unprocessed.hist(figsize=(20,20));
```

```python
df_unprocessed.columns
```

```python
import matplotlib.pyplot as plt

delay_types = ['CARRIER_DELAY', 'WEATHER_DELAY', 'NAS_DELAY', 'SECURITY_DELAY']
delay_counts = df_unprocessed[df_unprocessed['DEP_DEL15'] == 1][delay_types].sum()
delay_counts.plot(kind='bar')
plt.xlabel('Delay Types')
plt.ylabel('Count')
plt.title('Delayed Flights by Delay Types')
plt.show()


# # plot these columns 'CARRIER_DELAY','WEATHER_DELAY', 'NAS_DELAY', 'SECURITY_DELAY' with ARR_DEL15
# fig, ax = plt.subplots(2, 2, figsize=(20, 10))
# sns.scatterplot(x='CARRIER_DELAY', y='ARR_DEL15', data=df_unprocessed, ax=ax[0, 0])
# sns.scatterplot(x='WEATHER_DELAY', y='ARR_DEL15', data=df_unprocessed, ax=ax[0, 1])
# sns.scatterplot(x='NAS_DELAY', y='ARR_DEL15', data=df_unprocessed, ax=ax[1, 0])
# sns.scatterplot(x='SECURITY_DELAY', y='ARR_DEL15', data=df_unprocessed, ax=ax[1, 1]);

# # plot these columns 'CARRIER_DELAY','WEATHER_DELAY', 'NAS_DELAY', 'SECURITY_DELAY' with CANCELLED
# fig, ax = plt.subplots(2, 2, figsize=(20, 10))
# sns.scatterplot(x='CARRIER_DELAY', y='CANCELLED', data=df_unprocessed, ax=ax[0, 0])
# sns.scatterplot(x='WEATHER_DELAY', y='CANCELLED', data=df_unprocessed, ax=ax[0, 1])
# sns.scatterplot(x='NAS_DELAY', y='CANCELLED', data=df_unprocessed, ax=ax[1, 0])
# sns.scatterplot(x='SECURITY_DELAY', y='CANCELLED', data=df_unprocessed, ax=ax[1, 1]);
```

```python
# correlation chart to DEP_DEL15 with seaborn
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
sns.heatmap(df.corr()[['DEP_DEL15']], annot=True, cmap='coolwarm')
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
# Split the CRS_DEP_TIME column by DEP_DEL15 with dark red palette
sns.displot(df, x='CRS_DEP_TIME', hue='DEP_DEL15', palette='Reds', bins=24)
# sns.displot(df, x='CRS_DEP_TIME', hue='DEP_DEL15', bins=24)
plt.show()
```
